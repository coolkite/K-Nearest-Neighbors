```python
import csv
data = {}
with open('data.csv', newline='') as csvfile:
    reader = csv.reader(csvfile)
    header = next(reader)
    for row in reader:
        pid = row[0]
        features = eval(row[1])
        is_intrusive = int(row[2])
        data[pid] = {'features': features, 'is_intrusive': is_intrusive}
```

```python
print(data)
```

```python
def predict_label(examples, features, k, label_key="is_intrusive"):
    nearestNeighbors = find_k_nearest_neighbors(examples, features, k)
    intrusiveOrNot = {0: 0, 1:0}
    for pid in nearestNeighbors:
        if examples[pid]["is_intrusive"] == 1:
            intrusiveOrNot[1] += 1
        else:
            intrusiveOrNot[0] += 1

    mode = max(intrusiveOrNot, key=lambda x: intrusiveOrNot[x]) 
    return mode
```

```python
import math
def find_k_nearest_neighbors(examples, features, k):
    distanceMap = {}
    for pid in examples:
        for idx, val in enumerate(examples[pid]['features']):
            if pid not in distanceMap:
                distanceMap[pid] = (features[idx] - val)**2
                continue
            distanceMap[pid] += (features[idx] - val)**2
        distanceMap[pid] = math.sqrt(distanceMap[pid])

    distanceList = sorted(distanceMap, key = distanceMap.get)
    return distanceList[:k]
```

```python
sampleFeature = [4.30936122, 4.28739283, 4.29680938, 4.33571647, 4.28774593]
k = 1

nearestNeighbors = find_k_nearest_neighbors(data, sampleFeature, k)
predictedLabel = predict_label(data, sampleFeature, k)

print("Nearest neighbors are: " , nearestNeighbors)
print("Predicted value is: " , predictedLabel)
```
